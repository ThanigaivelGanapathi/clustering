```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import BisectingKMeans
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
bisecting = BisectingKMeans(n_clusters=3, random_state=0)
bisecting.fit(X)
```

```python
y_bisecting = bisecting.fit_predict(X)
y_bisecting
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_bisecting
```

```python
supervised
```

```python
supervised.to_csv("BisectingKmeansCluster.csv")
```
