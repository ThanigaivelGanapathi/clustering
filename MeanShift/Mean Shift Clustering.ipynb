```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
meanshift = MeanShift(bandwidth = 15)
meanshift.fit(X)
```

```python
y_meanshift = meanshift.fit_predict(X)
y_meanshift
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_meanshift
```

```python
supervised
```

```python
supervised.to_csv("MeanShiftCluster.csv")
```
