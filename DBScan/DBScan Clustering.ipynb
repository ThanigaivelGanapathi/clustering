```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
dbscan = DBSCAN(eps=3, min_samples=2)
dbscan.fit(X)
```

```python
y_dbscan = dbscan.fit_predict(X)
y_dbscan
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_dbscan
```

```python
supervised
```

```python
supervised.to_csv("DbScanCluster.csv")
```
