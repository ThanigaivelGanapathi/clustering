```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import OPTICS
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
optics = OPTICS(min_samples=10)
optics.fit(X)
```

```python
y_optics = optics.fit_predict(X)
y_optics
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_optics
```

```python
supervised
```

```python
supervised.to_csv("OpticsCluster.csv")
```
