```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import Birch
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
birch = Birch()
birch.fit(X)
```

```python
y_birch = birch.fit_predict(X)
y_birch
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_birch
```

```python
supervised
```

```python
supervised.to_csv("BirchCluster.csv")
```

```python


chart = sns.lmplot(data = supervised, x= supervised.columns[3], y = supervised.columns[4], hue = supervised.columns[5],fit_reg = False, legend = True, legend_out = True)
```
