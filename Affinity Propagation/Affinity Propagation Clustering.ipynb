```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation
```

```python
dataset = pd.read_csv("Mall_Customers.csv")
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
affinity = AffinityPropagation(random_state = 42)
affinity.fit(X)
```

```python
y_affinity = affinity.fit_predict(X)
y_affinity
```

```python
supervised = dataset
```

```python
supervised["Classification_Group"] = y_affinity
```

```python
supervised
```

```python
supervised.to_csv("AffinityCluster.csv")
```
